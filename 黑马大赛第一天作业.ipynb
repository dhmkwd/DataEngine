### Action2: 统计全班的成绩
#### 班里有5名同学，现在需要你用Python来统计下这些人在语文、英语、数学中的平均成绩、最小成绩、最大成绩、方差、标准差。然后把这些人的总成绩排序，得出名次进行成绩输出（可以用numpy或pandas）

```python
import pandas as pd
from pandas import Series, DataFrame
data = {'姓名': ['张飞','关羽','刘备','典韦','许褚'],'语文': [68, 95, 98, 90,80], '数学': [65, 76, 86, 88, 90], '英语': [30, 98, 88, 77, 90]}
df1 = pd.DataFrame(data)
df1
```

```python
df1.describe()
```

```python
df1['总成绩'] = df1[['语文','数学','英语']].sum(axis = 1)
# df1['总成绩'] = df1[2:].sum(axis = 1)
df1
```

```python
df1.sort_values('总成绩', ascending=False)
```

### Action3: 对汽车质量数据进行统计

```python
df = pd.read_csv('data/car_complain.csv')
df
```

```python
df = df.drop('problem', axis = 1).join(df.problem.str.get_dummies(','))
df
```

```python
def f(x):
    x=x.replace('一汽-大众','一汽大众')
    return x
df['brand'] = df['brand'].apply(f)
result = df.groupby(['brand'])['id'].agg(['count'])
result
```

```python
df['type'] = df['type'].apply(f)
result = df.groupby(['type'])['id'].agg(['count'])
result
```
